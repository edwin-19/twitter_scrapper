## Clean data

```python
import spacy
import re
import time
nlp = spacy.load('en_core_web_sm')
```

```python
def clean_text(text):
    text = " ".join([n.lemma_ if not n.is_stop else "" for n in text])
    text = str(text)
    text = text.lower()
    text = text.replace('{html}',"") 
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', text)
    rem_url = re.sub(r'http\S+', '',cleantext)
    rem_num = re.sub('[0-9]+', '', rem_url)
    clean_text = re.sub(r"[^a-zA-Z0-9]+", ' ', rem_num)
    
    
    # Normalize the word twealveans to twealvean
    clean_text = " ".join(
        ['twelvean' if t.text == 'twelveans' else t.text for t in nlp(clean_text)]
    )
    
    return clean_text
```

```python
cleand_text = []
start_time = time.time()
for index, doc in enumerate(nlp.pipe(df['tweet'].astype('unicode').values, batch_size=50, n_threads=3)):
    if doc.is_parsed:
        cleand_text.append(clean_text(doc))
                
end_time = time.time() - start_time
print('Time took to clean data {}'.format(end_time))
```

# EDA

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
def generate_word_cloud(text, is_save=False):
    # adding movie script specific stopwords
    stopwords = set(STOPWORDS)
    stopwords.add("https")
    stopwords.add("xa0")
    stopwords.add("xa0'")
    stopwords.add("bitly")
    stopwords.add("bit")
    stopwords.add("ly")
    stopwords.add("twitter")
    stopwords.add("pic")

    wordcloud = WordCloud(
        background_color='black',
        width=1000,
        height=500,
        stopwords=stopwords
    ).generate(text)
    
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.rcParams['figure.figsize'] = [10, 10]
    
    if is_save:
        wordcloud.to_file('cloud.png')
```

```python
generate_word_cloud(str(cleand_text), is_save=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
import time
import numpy as np 
```

```python
# Get term frequency
start_time = time.time()
tfidf = TfidfVectorizer(
    min_df=200, max_features=10000, strip_accents = 'unicode',
    analyzer = 'word', ngram_range = (1,1), use_idf = 1,
    smooth_idf = 1,sublinear_tf = True, stop_words = 'english'
)
tfidf.fit(cleand_text)
x = tfidf.fit_transform(cleand_text)
features = np.array(tfidf.get_feature_names())
print("Total time to compute unigrams",time.time()- start_time)
```

```python
pd.DataFrame(
    tfidf.idf_, index=tfidf.get_feature_names(), columns=['idf']
)
```

```python
pd.DataFrame(
    x.todense(), columns=tfidf.get_feature_names()
)
```
